```python
import pandas as pd
import numpy as np
import scanpy as sc
import matplotlib.pyplot as plt
import os
import sys
import scipy.sparse
```

# 151508

```python
# Set the section ID and paths
section_id = '151508'
input_dir = os.path.join('..', 'GraphST_part', section_id)

# Load the data
adata = sc.read_visium(path=input_dir, 
                      count_file=section_id+'_filtered_feature_bc_matrix.h5')
adata.var_names_make_unique()

# Print basic information about the loaded data
print(f"Loaded data shape: {adata.shape}")
print(f"Number of genes: {adata.n_vars}")
print(f"Number of spots: {adata.n_obs}")
```

```python
# 检查数据类型和基本统计信息
print("Data type:", type(adata.X))
print("Data shape:", adata.X.shape)
print("Min value:", adata.X.min())
print("Max value:", adata.X.max())
print("Mean value:", adata.X.mean())

# 如果是稀疏矩阵，转换为密集矩阵查看
if scipy.sparse.issparse(adata.X):
    print("\nSparse matrix info:")
    print("Number of non-zero elements:", adata.X.nnz)
    print("Sparsity: {:.2f}%".format(100 * (1 - adata.X.nnz / (adata.X.shape[0] * adata.X.shape[1]))))
```

```python
# 替代方案：使用方差来选择高变基因
# 1. 标准化和对数转换
print("Performing normalization and log transformation...")
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# 2. 手动计算基因变异度
print("Calculating gene variability...")
if scipy.sparse.issparse(adata.X):
    variances = adata.X.power(2).mean(0).A1 - np.power(adata.X.mean(0).A1, 2)
else:
    variances = np.var(adata.X, axis=0)

# 3. 选择前3000个高变基因
print("Selecting highly variable genes...")
n_top = 3000
top_genes_idx = np.argsort(variances)[-n_top:]
adata.var['highly_variable'] = False
adata.var.iloc[top_genes_idx] = True

print(f"\nNumber of highly variable genes: {sum(adata.var.highly_variable)}")
```

```python
# 读取注释文件
Ann_df = pd.read_csv(os.path.join('..', 'GraphST_part', '151508', '151508_layer_annotations.csv'))

# 将 barcode 列设置为索引
Ann_df.set_index('barcode', inplace=True)

# 找到共同的 barcodes
common_barcodes = list(set(adata.obs_names).intersection(set(Ann_df.index)))
adata = adata[common_barcodes].copy()

# 直接使用 'layer' 列（因为这是原始列名）
adata.obs['Ground Truth'] = Ann_df.loc[adata.obs_names, 'layer']

# 验证添加是否成功
print("\nFirst few rows of adata annotations:")
print(adata.obs['Ground Truth'].head())
print("\nUnique layers:", adata.obs['Ground Truth'].unique())
print("Number of spots per layer:")
print(adata.obs['Ground Truth'].value_counts())
```

```python
plt.rcParams["figure.figsize"] = (3, 3)
sc.pl.spatial(adata, img_key="hires", color=["Ground Truth"])
```

```python
STAGATE.Cal_Spatial_Net(adata, rad_cutoff=150)
STAGATE.Stats_Spatial_Net(adata)
```

```python
adata = STAGATE.train_STAGATE(adata, alpha=0)
```

```python
# 1. 检查 STAGATE 嵌入是否存在
print("Available embeddings:", list(adata.obsm.keys()))

sc.pp.neighbors(adata, use_rep='STAGATE')
sc.tl.umap(adata)
adata = STAGATE.mclust_R(adata, used_obsm='STAGATE', num_cluster=7)
```

```python
obs_df = adata.obs.dropna()
ARI = adjusted_rand_score(obs_df['mclust'], obs_df['Ground Truth'])
# 计算 NMI
NMI = normalized_mutual_info_score(obs_df['mclust'], obs_df['Ground Truth'])

# 打印结果
print('Adjusted rand index (ARI) = %.4f' % ARI)
print('Normalized mutual information (NMI) = %.4f' % NMI)
```

```python
plt.rcParams["figure.figsize"] = (3, 3)
sc.pl.umap(adata, color=["mclust", "Ground Truth"], title=['STAGATE (NMI=%.2f)'%NMI, "Ground Truth"])
```

```python
plt.rcParams["figure.figsize"] = (3, 3)
sc.pl.spatial(adata, color=["mclust", "Ground Truth"], title=['STAGATE (NMI=%.2f)'%NMI, "Ground Truth"])
```
